3.create legends in plots.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x), '-b', label='Sine')
plt.plot(x, np.cos(x), '--r', label='Cosine')

plt.legend()
```

4.Realize a vector t = [−10, 10] with increment 0.01 as an array.

```python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(-10,10,0.01)
plt.plot(t)
plt.xlabel("X - axis") 
plt.ylabel('Y - axis') 
plt.show()
```

5.Implement and plot the functions

• f(t) = cost

• f(t) = cost cos 5t + cos 5t

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 10, 0.01)
 
f1 = np.cos(t)
f2 = np.cos(10*t) 
f = f1*f2+f2

# plotting in subplot
plt.figure(figsize=(10, 10))

plt.subplot(3,1,1)
plt.plot(t,f1)
plt.title("cos(t) signal ") 


plt.subplot(3,1,2)
plt.plot(t,f2)
plt.title("cos(5t) signal") 

plt.subplot(3,1,3)
plt.plot(t,f)
plt.title("cos(t)*cos(5*t)+cos(5*t)") 

plt.show() # showing the figure
```
