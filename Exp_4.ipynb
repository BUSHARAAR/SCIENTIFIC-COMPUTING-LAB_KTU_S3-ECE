1. **Realize the functions sin t, cost, sinht and cosht for the vector t = [0, 10] with increment 0.01**

```python
import matplotlib.pyplot as plt 
import numpy as np
# Get x values of the sine wave 
t = np.linspace(0, 10,100)
y = np.sin(t)
z = np.cos(t)
m = np.sinh(t)
n= np.cosh(t)

fig = plt.figure(figsize=(10, 10))

plt.subplot(2,2,1)
plt.plot(t,y)
plt.title('Sin wave') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')


plt.subplot(2,2,2)
plt.plot(t,z)
plt.title('Cos wave') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')

plt.subplot(2,2,3)
plt.plot(t,m)
plt.title('sinht wave') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')

plt.subplot(2,2,4)
plt.plot(t,n)
plt.title('cosh wave') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')

```

2. **Compute the first and second derivatives of  sin t, cost, sinht and cosht functions using built in tools such as grad.**

```python
import matplotlib.pyplot as plt 
import numpy as np
# Get x values of the sine wave 
t = np.arange(0, 10, 0.01)
y = np.sin(t)
z = np.cos(t)


fdy1= np.gradient(y) # calculating first derivative of sine
fdy2= np.gradient(fdy1)	# calculating second derivative  sine


fdz1= np.gradient(z) # calculating first derivative cos
fdz2= np.gradient(fdz1)	# calculating second derivative cos

# plotting sine and cos signals in subplot
fig = plt.figure(figsize=(10, 10))

plt.subplot(2,3,1)
plt.plot(t,y)
plt.title('Sin wave') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')


plt.subplot(2,3,2)
plt.plot(t,fdy1)
plt.title('first derivative of sine wave') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')

plt.subplot(2,3,3)
plt.plot(t,fdy2)
plt.title('second derivative  sine wave') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')

plt.subplot(2,3,4)
plt.plot(t,z)
plt.title('cos wave') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')


plt.subplot(2,3,5)
plt.plot(t,fdz1)
plt.title('first derivative of cos wave') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')

plt.subplot(2,3,6)
plt.plot(t,fdz2)
plt.title('second derivative  cos') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')
```

```python
import matplotlib.pyplot as plt 
import numpy as np
# Get x values of the sine wave 
t = np.arange(0, 10, 0.01)

m = np.sinh(t)
n= np.cosh(t)


fdm1= np.gradient(m) # calculating first derivative of sinh(t)
fdm2= np.gradient(fdm1)	# calculating second derivative sinh(t)


fdn1= np.gradient(n) # calculating first derivative cosh(t)
fdn2= np.gradient(fdn1)	# calculating second derivative cosh(t)

fig = plt.figure(figsize=(10, 10))

plt.subplot(2,3,1)
plt.plot(t,m)
plt.title('sinh(t)') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')


plt.subplot(2,3,2)
plt.plot(t,fdm1)
plt.title('first derivative of sinh(t)') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')

plt.subplot(2,3,3)
plt.plot(t,fdm2)
plt.title('second derivative  sinh(t)') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')

plt.subplot(2,3,4)
plt.plot(t,n)
plt.title('cosh(t)') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')


plt.subplot(2,3,5)
plt.plot(t,fdn1)
plt.title('first derivative of cosh(t)') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')

plt.subplot(2,3,6)
plt.plot(t,fdn2)
plt.title('second derivative  cosh(t)') 
plt.xlabel('Time') 
plt.ylabel('Amplitude')
```

**3.Realize the function f(t) = 4t^2 + 3 and plot it for the vector t = [−5, 5] with increment 0.01**

```python
# realizing the function
import matplotlib.pyplot as plt 
import numpy as np
t = np.arange(-5, 5, .01); #assigning the vector t ans 
fn = (4*t**2)+3
plt.plot(t,fn)

plt.title("f(t)=4t^2 + 3") 
plt.xlabel("t") 
plt.ylabel("f(t)")
#show()
```

***4. Use general integration tool to compute intgral (f(t) dt) in [-2,2] , where f(t) = 4t^2 + 3 ***

```python
from scipy.integrate import quad #  quad is used for General purpose integration

def f(t):
  return 4*t**2+3

int, err = quad(f, -2, 2)

print("value of integral is : ",int) 
print("estimate of the absolute error is : ",err)
```

```python
from scipy.integrate import quad 

f= lambda x:4*x**2+3 # A lambda function can take any number of arguments, but can only have one expression.
i= quad(f, -2, 2)

print (i)
```

f(x) = e^{-x^2} integrated over a range of 0 and 1.

```python
import scipy.integrate
import numpy as np

f= lambda x:np.exp(-x**2)
i = scipy.integrate.quad(f, 0, 1)

print (i)
```

**5. Compute integration of 1 / √2π ( e ^ − x^2)/2dx ) in
[0,∞]**

```python
import scipy.integrate
import numpy as np

f= lambda x:1/(np.sqrt(2)*np.pi)*(np.exp(-x**2)/2)
i = scipy.integrate.quad(f, 0, np.inf)

print (i)
```
