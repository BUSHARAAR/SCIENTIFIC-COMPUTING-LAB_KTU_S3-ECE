**1. Solve the first order differential equation dx / dt + 2x = 0, with the initial condition x(0) = 1**

The following code solves the above equation using the following stepts:

import Numpy, Scipy, and Matplotlib packages

Create function model based on the equation to be solved

Create an array of time points (choose arbitrarily for time being)

Set initial condition on x(0). You can try different initial condition.

Call odeint function to numerically calculate x(t) and store the result in 
user-defined variable solution.

Plot the solution.

```python
import numpy as np
from scipy.integrate import odeint 
import matplotlib.pyplot as plt

# function that returns dx/dt 
def model(x,t):
  dxdt = -2 * x 
  return dxdt

# initial condition 
x0 = 1
# time points

t = np.linspace(0,20,50)
#print(t)

# solve ODE
x = odeint(model,x0,t) # model fn, initial condition, time point
#print(x)

# plot results 
plt.figure(figsize=(5, 5))
plt.plot(t,x)
plt.title("FIrst order Differential equation response") 
plt.xlabel('time')
plt.ylabel('x') 
plt.legend() 
plt.show() 
```

**2. Solve the second order differential equation d^2x / dt^2 + 2 dx/dt + 2x = e^−t**

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.integrate import odeint


def model(X, t):
  dx2dt = [X[1], -2*X[1] - 2*X[0] + np.exp(-t)] 
  return dx2dt

X0 = [0,1] #initial condition

t = np.linspace(0, 10, 100) 
Xs = odeint(model, X0, t)
ys = Xs[:,0]
ys1= Xs[:,1]

plt.figure(figsize=(5, 5))
plt.xlabel("t")
plt.ylabel("y")
plt.title("second order diffrential equation response")
plt.plot(t,ys,label='ys')
plt.plot(t,ys1, label='ys1')
plt.legend() 
```

**3. Solve for the current transient through an RC network (with RC = 3) that is driven by 
 
*   5 V DC , and plot the solutions

*   Using the signal 5e^-tU(t)

```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0,10,50)
Vs= 5
R = 3000 #R Value
C = 1000 * 10 ** (-6) #C value

Vt = Vs*(1-np.exp(-t/R*C))

i = (Vs/R)*np.exp((-1/(R*C))*t)

plt.plot(t,i,label="Current") # Current graph
plt.plot(t,Vt,label="Voltage") # voltage graph

plt.xlabel('Time (s)') 
plt.title('RC circuit') 
plt.legend()
plt.show()

tau= R*C
print(tau)
```

```python
#Using the signal 5e^-tU(t)
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0,10,50)
Vs= 5 * np.exp(-1*t)
R = 3000 #R Value
C = 1000 * 10 ** (-6) #C value

Vt = Vs*(1-np.exp(-t/R*C))

i = (Vs/R)*np.exp((-1/(R*C))*t)
plt.figure(figsize=(5, 5))

plt.plot(t,i,'+',label="Current") # Current graph
plt.plot(t,Vt,'*',label="Voltage") # voltage graph

plt.xlabel('Time (s)') 
plt.title('RC circuit') 
plt.legend()
plt.show()

tau= R*C
print(tau)
```

4. Solve the current transient through a series RLC circuit with R = 1Ω, L = 1 mH and C = 1 μF that is driven by

• 5 V DC

. and Using the signal 5e^-tU(t) and plot the solutions.

```python
#importing the library

import numpy as np
import matplotlib.pyplot as plt
L=1 # Inductor
R=1 # value of resistor
C=1*10**-3 #value of capacitor
tow=2*(L/R) #time constant
t = np.arange(0, 10, 0.01)
V1=5 
V2=5*np.exp(-1*t)

w=np.sqrt((1/(L*C)-((R*R)/(4*(L*L))))) # resonant frequency
Resp1 = V1*np.exp(-1*(t)/tow)*np.cos(w*t) # response with v = 5V
Resp2 = V2*np.exp(-1*(t)/tow)*np.cos(w*t) # response with  v =5e^(-t)U(t) 

#Plotting the values

plt.figure(figsize=(5, 5))

plt.subplot(2,1,1)
plt.plot(t,Resp1)
plt.title("Transient response for V =5V")
plt.xlabel("t")

plt.subplot(2,1,2)
plt.plot(t,Resp2)
plt.title("Transient response for V=5e^(-t)U(t)")
plt.xlabel("t")
plt.show()
```
